# Visualizaciones de soluciones de ecuaciones diferenciales

# El modelo SIR

```python
import numpy as np
from scipy.integrate import odeint
import matplotlib.pyplot as plt
```

##### Parámetros del modelo 

```python
beta, gama= 0.4,0.7
tmax=365
t=np.linspace(0,tmax,tmax*10)
```

##### Definición del lado derecho del modelo

```python
def modSIR(y,t,beta,gama):
    S, I, R =y
    dSdt= -beta*S*I
    dIdt= beta*S*I-gama*I
    dRdt= gama*I
    return dSdt,dIdt,dRdt
```

##### Condiciones iniciales

```python
S0, I0, R0 = 25,15,0
y0 = S0, I0, R0
#
solucion=odeint(modSIR,y0,t,args=(beta,gama))
S,I,R=solucion.T
```

### Gráfica de la solución

```python
%matplotlib inline
fig= plt.figure()
plt.plot(t,S,'b',label="Suceptibles")
plt.plot(t,I,'r',label="Infecciosos")
plt.plot(t,R,'g',label="Recuperados")
plt.xlabel("Tiempo")
plt.ylabel("Número")
plt.xlim(0,10)
#plt.ylim(0,1.2)
legend =fig.legend()
legend.get_frame().set_alpha(0.5)
```
